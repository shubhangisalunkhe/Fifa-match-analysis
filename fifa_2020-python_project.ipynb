# Data Preprocessing:
Dropping some useless columns

```python
useless_columns = ['dob','sofifa_id', 'player_url', 'long_name', 'body_type', 'real_face', 'loaned_from', 'nation_position', 'nation_jersey_number']
```

```python
fifa = fifa.drop(useless_columns, axis=1)
```

```python
fifa.head()
```

# Player's Position

```python
fifa[['short_name','player_positions']]
```

```python
new_player_positions = fifa['player_positions'].str.get_dummies(sep = ', ').add_prefix('Position')
new_player_positions.head()
```

```python
fifa=pd.concat([fifa,new_player_positions], axis=1)
```

```python
fifa.head()
```

# Positioning Columns ratings

# Cleaning, Processing and Assigning the new attributes to columns listed below.

```python
columns = ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram',
       'lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb',
       'lcb', 'cb', 'rcb', 'rb']
```

```python
fifa[columns].head()
```

# Now, lets omit the '+' sign

```python
for col in columns:
    fifa[col] = fifa[col].str.split('+', n=1, expand = True)[0]
    fifa[columns]
```

```python
fifa[columns]
```

# Deleting the Nan values, lets replace Nan with 0

```python
fifa[columns] = fifa[columns].fillna(0)
```

# Converting columns into int here

```python
fifa[columns] = fifa[columns].astype(int)
```

```python
fifa[columns]
```

# Filling missing values
Let's fill "dribbling","defending", "physic", "passing", "shooting" and "pace" missing values of these columns by median

```python
columns = ["dribbling", "defending", "physic", "passing", "shooting", "pace"]
fifa[columns]
```

```python
fifa[columns].isnull().sum()
```

```python
for col in columns:
     fifa[col] = fifa[col].fillna(fifa[col].median())
fifa[columns]
```

```python
fifa = fifa.fillna(0)
```

```python
fifa.isnull().sum()
```

```python
plt.figure(figsize=(5,7))
plt.hist(data=fifa,x="age")
plt.title('Histogram of Players ages')
plt.xlabel('age')
plt.show()
```
